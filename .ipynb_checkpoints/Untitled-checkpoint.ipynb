```python
from scipy import stats
import numpy as np

# Calcular el Z-score para la cantidad
df['z_score'] = np.abs(stats.zscore(df['Quantity']))

# Identificar anomal√≠as
anomalies_z_score = df[df['z_score'] > 3]
```

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1

# Identificar anomal√≠as
anomalies_iqr = df[(df['Quantity'] < (Q1 - 1.5 * IQR)) | (df['Quantity'] > (Q3 + 1.5 * IQR))]
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.05)  # Suponiendo que el 5% de los datos son anomal√≠as
df['anomaly'] = model.fit_predict(df[['Quantity']])

# Filtrar las anomal√≠as
anomalies_isolation_forest = df[df['anomaly'] == -1]
```

```python
# Definir el umbral para el Z-score
threshold = 3

# Identificar las transacciones an√≥malas
anomalies = df[(df['Z_score'] > threshold) | (df['Z_score'] < -threshold)]

# Mostrar las anomal√≠as
print(anomalies)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest

# Normalizar las cantidades
scaler = StandardScaler()
df['Quantity_scaled'] = scaler.fit_transform(df[['Quantity']])

# Ajustar el modelo Isolation Forest
model = IsolationForest(contamination=0.05)
df['anomaly'] = model.fit_predict(df[['Quantity_scaled']])

# Filtrar las anomal√≠as
anomalies_isolation_forest = df[df['anomaly'] == -1]

# Visualizar las anomal√≠as
plt.figure(figsize=(10, 6))
plt.scatter(df['InvoiceDate'], df['Quantity'], c='blue', label='Normal')
plt.scatter(anomalies_isolation_forest['InvoiceDate'], anomalies_isolation_forest['Quantity'], c='red', label='Anomaly')
plt.xlabel('Invoice Date')
plt.ylabel('Quantity')
plt.title('Anomal√≠as en Cantidades de Productos')
plt.legend()
plt.show()

```
